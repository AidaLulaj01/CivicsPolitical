```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Connect with Postgres
engine = create_engine('postgresql://postgres:Constantina06*@localhost:5432/Civics_Political')
```

```python
#Join all the tables created on the DB
df = pd.read_sql('select * from voter_info vi join voter_choice vc on vi.case_id = vc.case_id join candidate_info ci on vc.candidate_id = ci.candidate_id join ideological_leaning il on il.ideoId = vi.ideoId join district_info di on vi.zipcode = di.zipcode order by vi.case_id desc', engine.connect())
```
